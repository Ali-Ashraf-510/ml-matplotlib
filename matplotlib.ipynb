```python
df_houses = pd.read_csv('housing.csv')
```

```python
df_houses.head()
```

```python
# تعيين حجم الشكل (Figure) ليكون بعرض 10 وارتفاع 6
plt.figure(figsize=(10, 6))

# إنشاء مخطط انتشار (scatter plot) لتوزيع البيوت حسب خط الطول (longitude) وخط العرض (latitude)
# alpha: الشفافية 0.5
# s: حجم النقاط بناءً على عدد السكان (population) مقسوم على 100
# c: لون النقاط بناءً على متوسط قيمة المنزل (median_house_value)
# cmap: تعيين خريطة الألوان إلى 'jet' لتغيير الألوان تلقائيًا حسب القيمة
sc = plt.scatter(df_houses['longitude'], df_houses['latitude'],
                 alpha=0.5,
                 s=df_houses['population'] / 80,
                 c=df_houses['median_house_value'],
                 cmap='viridis',  # اختيار خريطة ألوان 'viridis' لمظهر أكثر احترافية
                 edgecolor='k',  # إضافة حواف سوداء للنقاط لزيادة التباين
                 linewidth=0.5,  # تقليل سماكة حواف النقاط
                 label="population")

# إضافة شريط الألوان للمخطط لتمثيل قيم متوسط سعر المنزل
plt.colorbar(sc)

# تسمية المحور السيني بخط العرض (latitude) مع ضبط الحجم واللون
plt.xlabel('latitude', fontsize=14, color='r')

# تسمية المحور الصادي بخط الطول (longitude) مع ضبط الحجم واللون
plt.ylabel('Longitude', fontsize=14, color='blue')
plt.yticks(fontsize=12)  

# إضافة عنوان للمخطط لتوضيح المحتوى
plt.title('longitude vs latitude', fontsize=14, color='r')

# عرض وسيلة إيضاح لتوضيح العنصر الممثل
plt.legend()
plt.grid(visible=True , linestyle='--' ,alpha = 0.5 , )
# عرض المخطط النهائي
plt.show()
```

```python
x = np.linspace(1, 5, 100) 
y = np.exp(x)  
z = np.log2(x)
```

```python
plt.plot(x, y, marker='o', color='r', markersize=2, label='exp(x)')
plt.plot(x, z, marker='+', color='g', markersize=2, label='log10(x)')

## name the axis and make a title
plt.title('x vs. y')

plt.xlabel('x')
plt.ylabel('y')

## legend
plt.legend()

## show the graph
plt.show()
```

```python
## twinx

## We want to optimize the above figure
fig, ax = plt.subplots()

ax.plot(x, y, color='blue', label='exp(x)')
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y', color='blue', fontsize=14)

## create another axis
ax2 = ax.twinx()
ax2.plot(x, z, color='red', label='log10(x)')

ax2.set_ylabel('z', color='red', fontsize=14)

plt.title('x vs. y & x vs. z', fontsize=14)
plt.show()
```

```python
# حساب عدد التكرارات لقيم المسافة من المحيط وتحديد الفئات
val = df_houses['ocean_proximity'].value_counts().values
ind = df_houses['ocean_proximity'].value_counts().index

# تعيين حجم الشكل ليكون بعرض 10 وارتفاع 6
plt.figure(figsize=(10, 6))

# رسم الرسم البياني الدائري مع تحسينات في الألوان والشفافية
plt.pie(x=val, 
        labels=ind, 
        autopct='%1.2f%%',  # عرض النسبة بدقة تصل إلى منزلتين عشريتين
        explode=[0.15] * len(ind),  # إضافة تأثير الانفصال (explode) لكل جزء
         startangle=140,  # بدء الرسم بزاوية 140 لزيادة جاذبية العرض
        colors=plt.cm.Paired.colors,  # استخدام ألوان متناسقة من خريطة الألوان 'Paired'
        wedgeprops={'edgecolor': 'black', 'linewidth': 1})  # إضافة حواف سوداء لكل جزء

# إضافة عنوان للرسم البياني
plt.title("Distribution of Houses by Ocean Proximity", fontsize=16, color='blue')

# عرض الرسم البياني النهائي
plt.show()
```
